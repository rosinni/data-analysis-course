### 4) Limpieza: aplicar las decisiones

Ahora sí, implementen las decisiones del Paso 3. Recuerden:

- Primero eliminen duplicados
- Después manejen nulos
- Después conviertan tipos

```python
# Limpieza de pacientes
df_pacientes_limpio = df_pacientes.copy()
# Tu código acá
# 1. Primero: Eliminar duplicados
df_pacientes = df_pacientes.drop_duplicates(subset=['dni'], keep='first')

# 2. Después: Manejar nulos
df_pacientes = df_pacientes.dropna(subset=['dni'])
df_pacientes['ciudad'] = df_pacientes['ciudad'].fillna("Desconocida")
df_pacientes['obra_social'] = df_pacientes['obra_social'].fillna("Particular")

# 3. Después: Convertir tipos
df_pacientes['dni'] = pd.to_numeric(df_pacientes['dni'], errors='coerce')
df_pacientes['fecha_nacimiento'] = pd.to_datetime(df_pacientes['fecha_nacimiento'], errors='coerce')

print(f"✅ 'pacientes' finalizado ({df_pacientes.shape[0]} filas).")
print("-" * 50)




# Limpieza de médicos
df_medicos_limpio = df_medicos.copy()
# Tu código acá
# 1. Primero: Eliminar duplicados
df_medicos = df_medicos.drop_duplicates(subset=['matricula'], keep='first')

# 2. Después: Manejar nulos
df_medicos['matricula'] = df_medicos['matricula'].fillna("S/M")

# 3. Después: Convertir tipos
mapeo_activo = {'True': True, 'False': False, True: True, False: False}
df_medicos['activo'] = df_medicos['activo'].map(mapeo_activo)
df_medicos['activo'] = df_medicos['activo'].fillna(True)  # Relleno lógico post-mapeo

print(f"✅ 'medicos' finalizado ({df_medicos.shape[0]} filas).")
print("-" * 50)


# Limpieza de turnos
df_turnos_limpio = df_turnos.copy()
# Tu código acá
# 1. Primero: Eliminar duplicados (si existieran registros 100% idénticos)
df_turnos = df_turnos.drop_duplicates(keep='first')

# 2. Después: Manejar nulos en costo usando map y la función limpiadora
def limpiar_costo(valor):
    if pd.isnull(valor):
        return valor
    texto = str(valor).strip()
    if texto in ["Gratis", "N/A", "gratis"]:
        return 0.0
    texto = texto.replace("$", "").replace(".", "").replace(",", ".")
    try:
        return float(texto)
    except:
        return 0.0

df_turnos['costo'] = df_turnos['costo'].map(limpiar_costo)
df_turnos['costo'] = df_turnos['costo'].fillna(0.0)

# 3. Después: Convertir tipos (Casteo a datetime unificado)
df_turnos['fecha_completa'] = pd.to_datetime(df_turnos['fecha'] + ' ' + df_turnos['hora'], errors='coerce')

print(f"✅ 'turnos' finalizado ({df_turnos.shape[0]} filas).")
```

|> **Por qué `.copy()`:** cuando hacen `df_pacientes_limpio = df_pacientes`, no están creando una copia — están apuntando al mismo objeto. Cualquier cambio en `df_pacientes_limpio` modifica también `df_pacientes`. `.copy()` crea una copia independiente, así conservan los datos originales para comparar.
